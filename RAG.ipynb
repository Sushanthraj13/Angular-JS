```python
import torch
import transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel



#load the flcon model and tokenizer
model = "tiiuae/falcon-7b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model)
falcon_pipeline = transformers.pipeline("text-generation",
                                        model=model,
                                        tokenizer=tokenizer,
                                        torch_dtype=torch.bfloat16,
                                        trust_remote_code=True,
                                        device_map="auto"
                                        )



#Example document corpus
documents = [
    "Artificial intelligence is the simulation of human intelligence in machines.",
    "Machine learning is a subset of AI that focuses on data and algorithm.",
    "Deep learning is a further usbset of machine learning that uses neural networks.",
    "Natural Langugae Procssing enables machines to unerstand and respond to human language."
]
```

```python
vectorizer = TfidfVectorizer()
doc_vectors = vectorizer.fit_transform(documents)
```

```python
def retrieve_documents(query,top_n=2):
  query_vector = vectorizer.transform([query])
  similarities = linear_kernel(query_vector, doc_vectors).flatten()
  top_indices = similarities.argsort()[-top_n:][::-1]
  return [documents[i] for i in top_indices]
```

```python
def generate_response(query):
  retrieved_docs = retrieve_documents(query)
  context = " ".join(retrieved_docs)
  full_prompt = f"Context: {context}\nQuery: {query}\nAnswer:"
  generated_text = falcon_pipeline(full_prompt, max_length=100)
  return generated_text[0]['generated_text']
```

```python
if __name__ == "__main__":
  query = "What is Deep Learning?"
  response = generate_response(query)
  print("Response:", response)
```
